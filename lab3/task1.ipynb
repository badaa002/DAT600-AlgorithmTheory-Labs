```python
from collections import defaultdict
# inspiration for code found here: https://www.geeksforgeeks.org/python-program-for-topological-sorting/ 
class Graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def add_edge(self, u, v):
        # if the node u isnt in the graph then add it
        if u not in self.graph:
            self.graph[u] = []
        # if the node v isnt in the graph then add it
        if v not in self.graph:
            self.graph[v] = []
        # create edge from u to v
        self.graph[u].append(v)

    def top_util(self, node, visited, stack):
        # add node to the visited set
        visited.add(node)
        # dfs of the nodes
        for i in self.graph[node]:
            # if the index isnt visited yet (visited[i] equal to False)
            # then it will run the function recursively untill all the neighbours of the node are also visited
            if i not in visited:
                self.top_util(i, visited, stack)
        # add node to stack
        stack.append(node)

    def top_sort(self):
        # create a set for the visited nodes to avoid duplicates
        visited = set()
        # stack to store sorted nodes
        stack = []
        # loop through the nodes in the graph
        for i in self.graph:
            # if a node isnt visited yet, do dfs to perform the topological sorting
            if i not in visited:
                self.top_util(i, visited, stack)
        print(stack[::-1])
        # return the top. sorted stack in reverse order
        return stack[::-1]

# addition for d)
    def make_dag(self):
        # get the top. order created in the top_sort function
        order = self.top_sort()
        # create new graph for the dag
        dag = Graph()
        # create a set for the visited nodes
        visited = set()
        # loop through the nodes in top. order
        for i in order:
            # add nodes to visited set
            visited.add(i)
            # loop through the neighbours of the specific node
            for j in self.graph[i]:
                # loop through the neighbours of the specific node
                if j in visited:
                    print("Removed: " + "("+ i +","+ j + ")")
                    continue
                # create edge from node to neighbour
                dag.add_edge(i, j)
        print("")
        print("Edges in DAG:")
        for i in dag.graph:
            for j in dag.graph[i]:
                print(i, "->", j)
        

```
