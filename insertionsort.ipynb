```python
def insertion_sort(array):
    steps = 0
    for i in range(1, len(array)):
        steps+=2
        number = array[i]
        i -= 1
        while i>0 and number< array[i]:
            array[i+1]=array[i]
            i-=1
            steps+=2
        array[i+1]=number
        steps +=1
    return steps
```

```python
array = [1,4,6,2,5,9,6,7,88,9,7]
steps = insertion_sort(array)
print(array)
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [16,4]
from IPython.display import clear_output

array = [1,4,6,2,5,7,9,4,1,5,7,9,7,8]
points=10
x=[]
y=[]
ref=[]
c=[]

for i in range(0,points):
    clear_output(wait=True)
    steps=insertion_sort(array) 
    array=array*2
    x.append(len(array))
    y.append(steps)    
    c.append(steps/len(array)**2)
    
for j in x:
    ref.append([j**2 *c[4]])
plt.plot(x,y, color="purple", label="Insertionsort time complexity")
plt.plot(x,ref, color="green", label= ("O(n^2)"))
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Steps')
plt.grid(True)
plt.legend()
plt.show()
```
