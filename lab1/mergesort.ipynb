```python
def mergesort(array):
    steps = 0
    if len(array)>1:
        #finds the halfwaypoint of the array, which is where it will be split
        halfway_point = len(array)//2
        #devides the array into rightside and leftside arrays
        rightside = array[halfway_point:]
        leftside = array[:halfway_point]
        #sorts each array
        steps+=mergesort(rightside)
        steps+=mergesort(leftside)
        # indexes for the arrays
        # keeps track in leftside array
        x = 0
        # keeps track in rightside array
        y = 0
        # keeps track in original array
        z = 0
        while x < len(leftside) and y < len(rightside):
            steps+=1
            if leftside[x]<rightside[y]:
                array[z] = leftside[x]
                x+=1
            else:
                array[z] = rightside[y]
                y+=1
            z+=1
        while x < len(leftside):
            array[z]=leftside[x]
            steps+=1
            z+=1
            x+=1    
        while y < len(rightside):
            array[z]=rightside[y]
            steps+=1
            z+=1
            y+=1
    return steps
```

```python
array = [1,4,6,2,5,9,6,7,88,9,0,7]
steps = mergesort(array)
print(array)
```

```python
# code inspired by the insertion sort code example on canvas: https://stavanger.instructure.com/courses/13723/files/1614576?module_item_id=316270
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['figure.figsize'] = [16,4]

from IPython.display import clear_output


array = [1,4,6,2,5,7,9,4,1,5,7,9,7,8]
points=10
x=[]
y=[]
reference=[]
c=[]

for _ in range(0,points):
    clear_output(wait=True)
    steps=mergesort(array) 
    array*=2
    x.append(len(array))
    y.append(steps)    
    c.append(steps/len(array)/np.log(len(array)))
    
for i in x:
    reference.append([i*np.log(i)*c[4]])

plt.plot(x, y, color="purple", label='Merge Sort Time Complexity')
plt.plot(x, reference, color="green", label='n*log(n)')
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Steps')
plt.grid(True)
plt.legend()
plt.show()
```
